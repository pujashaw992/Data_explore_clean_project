```python
lower_limit, upper_limit = dp.basesalary.quantile([0.05,0.995])  
lower_limit, upper_limit

#lets set the maximum considered percentile to be 99.5% and minimum to be 5%
```

```python
outliers = dp[(dp.basesalary>upper_limit) | (dp.basesalary<lower_limit)]
outliers

#lets check the outlier values we have in out dataset(using Or operator)
```

```python
dp3 = dp[(dp.basesalary<upper_limit) & (dp.basesalary>lower_limit)]

#lets remove the outliers and take out our outlier free dataset (using And operator)
```

```python
dp3
```

```python
Q1 = dp.basesalary.quantile(0.25)
Q3 = dp.basesalary.quantile(0.75)

#lets get the 1st quartile(Q1) and 3rd quartile(Q3) from our Value_lacs column
```

```python
Q1,Q3
```

```python
IQR = Q3-Q1  #lets find the IQR (Inter Quartile Range)

IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit,upper_limit

#lets calculate the upper and lower limit to be considered for outlier removal
```

```python
dp[(dp.basesalary<lower_limit)|(dp.basesalary>upper_limit)] #lets check the rows having outliers
```

```python
dp4 = dp[(dp.basesalary>lower_limit)&(dp.basesalary<upper_limit)] #removing outlier and taking final dataset
dp4
```

```python
upper_limit = dp.basesalary.mean() + 2*dp.basesalary.std()
lower_limit = dp.basesalary.mean() - 2*dp.basesalary.std()

upper_limit,lower_limit

#lets consider 3 standard deviation for outlier detection
```

```python
dp[(dp.basesalary>upper_limit) | (dp.basesalary<lower_limit)] #lets check the rows having value beyond 2 std
```

```python
dp_no_outlier_std_dev = dp[(dp.basesalary<upper_limit) & (dp.basesalary>lower_limit)]

dp_no_outlier_std_dev

#lets clean the outlier and get the clean dataset
```
